```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math

dataa=open("kmeans_data.txt","r")
d=dataa.read().split()

data=[]
for i in range(0,len(d),2):
  data.append([float(d[i]),float(d[i+1])])

ReducedData=[]
for i in data:
  ReducedData.append(i[0]*i[0]+i[1]*i[1])


mean1=ReducedData[0]
mean2=ReducedData[1]

for j in range(10):
  c1=[]
  c1Data=[]
  c2=[]
  c2Data=[]
  for i in range(len(ReducedData)):
    if np.linalg.norm(ReducedData[i]-mean1)<np.linalg.norm(ReducedData[i]-mean2):
      c1.append(i)
      c1Data.append(ReducedData[i])
    else:
      c2.append(i)
      c2Data.append(ReducedData[i])
  mean1=np.mean(c1Data)
  mean2=np.mean(c2Data)
# print(mean1,mean2)
x1=[data[i][0] for i in c1]
y1=[data[i][1] for i in c1]
x2=[data[i][0] for i in c2]
y2=[data[i][1] for i in c2]

plt.scatter(x1, y1, marker='o', color='r', label='Cluster 1')
plt.scatter(x2, y2, marker='o', color='g', label='Cluster 2')

plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('K-means')

plt.legend()

plt.show()



```
