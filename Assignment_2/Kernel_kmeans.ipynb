```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import random

dataa=open("kmeans_data.txt","r")
d=dataa.read().split()

data=[]
for i in range(0,len(d),2):
  data.append([float(d[i]),float(d[i+1])])
print(data)
```

```python
def RBF(x1,x2):
  x=np.array(x1)
  y=np.array(x2)
  z=x-y
  return math.exp(-0.1*((np.linalg.norm(z))**2))
print(RBF(data[0],data[1]))

```

```python

for j in range(10):
  random_data=random.randint(0, len(data))
  new_data=[]
  for i in data:
    new_data.append(RBF(data[random_data],i))
  mean1=new_data[0]
  mean2=new_data[1]

  for j in range(20):
    c1=[]
    c1Data=[]
    c2=[]
    c2Data=[]
    for i in range(len(new_data)):
      if np.linalg.norm(new_data[i]-mean1)<np.linalg.norm(new_data[i]-mean2):
        c1.append(i)
        c1Data.append(new_data[i])
      else:
        c2.append(i)
        c2Data.append(new_data[i])
    mean1=np.mean(c1Data)
    mean2=np.mean(c2Data)
  print(mean1,mean2)
  x1=[data[i][0] for i in c1]
  y1=[data[i][1] for i in c1]
  x2=[data[i][0] for i in c2]
  y2=[data[i][1] for i in c2]

  plt.scatter(x1, y1, marker='o', color='r', label='Cluster 1')
  plt.scatter(x2, y2, marker='o', color='g', label='Cluster 2')
  plt.scatter(data[random_data][0],data[random_data][1], marker='o', color='b', label='landmark')

  plt.xlabel('X-Axis')
  plt.ylabel('Y-Axis')
  plt.title('Simple Scatterplot')

  plt.legend()

  plt.show()
```
